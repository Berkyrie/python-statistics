```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.seed(40)
sample_size = 1000
original_data = np.random.exponential(scale=1.0, size=sample_size)
```

```python
plt.hist(original_data, bins=30, density=True)
plt.title("Original Data Distribution")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```

```python
def clt_simulation(data, sample_size, num_samples):
    sample_means = []
    for _ in range(num_samples):
        sample = np.random.choice(data, size=sample_size, replace=True)
        sample_means.append(np.mean(sample))
    return sample_means

num_samples = 1000
sample_means = clt_simulation(original_data, sample_size=50, num_samples=num_samples)
```

```python
plt.hist(sample_means, bins=30, density=True, alpha=0.6, label="Sample Means")
plt.title("Distribution of Sample Means (CLT)")
plt.xlabel("Sample Mean")
plt.ylabel("Frequency")
```

```python
# When we increased amount of sample or number of samples, in both senario distribution of sample means chart looks more like a normal distribution
```
